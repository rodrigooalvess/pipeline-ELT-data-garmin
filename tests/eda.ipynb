```python
import pandas as pd
import unicodedata
```

```python
path = "../data/landing/activity_21851453697.csv"
```

```python
df = pd.read_csv(path)
df.head()
```

```python
def normalized_columns(col):
    col = col.lower().strip()
    col = unicodedata.normalize('NFKC', col).encode("ascii", "ignore").decode("utf-8") # estudar isso aqui
    col = col.replace(" ", "_").replace(":", "")

    return col
```

```python
df_cols_normalized = df.copy()
df_cols_normalized.columns = [normalized_columns(col) for col in df.columns]
df_cols_normalized.head()
```

```python
df_cols_normalized.dropna(axis=1 ,how="all", inplace=True)
df_cols_normalized.head()
```
